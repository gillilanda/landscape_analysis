```python
### input .csv files with raster aspect data
### output rose plots of each .csv file 
### PATHS FOR MAC

import os
import csv
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go


csv_folder = r"/Volumes/TAM/tam/GIS_stuff/aspect_shit/csv"
```

```python
def csvReader(file):
    actualList = []
    with open(file) as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        next(csv_reader)
        
        for row in csv_reader:
            if int(row[0]) >0:
                actualList.append(int(row[0]))
            else:
                pass
            
    actualList.sort()
    return actualList  
```

```python
for file in os.listdir(csv_folder):
    if file.endswith(".csv"):
        try:
            basin_name = os.path.splitext(file)[0]
            actualList = csvReader(csv_folder + "/" + file)
            actualListRad = np.radians(actualList)

            fig = plt.figure()

            ax = fig.add_subplot(111, polar=True)
            c = ax.hist(actualListRad, bins=12, density=True)

            ax.set_theta_zero_location('N', offset=0)
            ax.set_theta_direction(-1)
            plt.title(basin_name)
            plt.savefig(basin_name + ".pdf")
        except:
            print("Could not read " + basin_name)
```
