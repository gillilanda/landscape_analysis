```python
### input multiple .dbf files with slope data
### output normalized area versus slope graphs

import sys
import collections
import csv
import os
import simpledbf
import matplotlib.pyplot as plt

dbf_folder = r"d:/tam/dbf_slope"
csv_folder = r"d:/tam/csv"
plt_folder = r"d:/tam/slope_plots"
```

```python
def convert_to_csv(dbf_file):
    db = simpledbf.Dbf5(dbf_file)
    db.to_csv(r"d:/tam/csv/" + basin_name + ".csv")
    csv_file = r"d:/tam/csv/" + basin_name + ".csv"
    return csv_file
```

```python
def csv_reader(file):
    # make a list to put the original elevation data
    og_slope = []
    
    # open & read .csv file 
    with open(file) as csvFile:
        csv_reader = csv.reader(csvFile, delimiter=',')
        # Skip header
        next(csv_reader)
        
        for row in csv_reader:
            if int(row[0]) > 0:
                og_slope.append(int(row[0]))
            else:
                pass
            
    # delete .csv file
    os.remove(csv_folder + "/" + basin_name + ".csv")  
        
    og_slope.sort()
    return og_slope
```

```python
def count_values(list1):
    slope_values = []
    slope_freq = []
    
    counter = collections.Counter(list1).items()

    for nums in counter:
        slope_values.append(nums[0])
        slope_freq.append(nums[1])
    
#    print(counter)
#    print("*****************************************")
#    print(slope_values, slope_freq)
    return slope_values, slope_freq
```

```python
def normalized_area(list2):
    norm_area = []
    max_slope = max(list2)
    min_slope = min(list2)
    
    for values in list2:
        normed_value = (float(values)- (min_slope))/((max_slope) - (min_slope))
        
        norm_area.append(normed_value)
        
    return norm_area
```

```python
# Loop through each file in dbf_folder & create graphs

for file in os.listdir(dbf_folder):
    if file.endswith(".dbf"):
#        try:
        basin_name = os.path.splitext(file)[0]
        csv_file = convert_to_csv(dbf_folder +"/"+ file)
        og_slopes = csv_reader(csv_file)
        slope_values, slope_freq = count_values(og_slopes)
        norm_area = normalized_area(slope_freq)

        # plot
        plt.figure() 
        plt.plot(slope_values, norm_area, label= basin_name)
        plt.xlabel('Normalized Area')
        plt.ylabel('Slope')
        plt.legend()
        plt.title('Normalized Slope Frequency of All TAM Basins')
        plt.savefig("ploty.pdf")
        plt.show
#        except:
#            print("Could not read " + basin_name)
                        
```
